```python
#Import dependencies
import matplotlib.pyplot as plt
import pandas as pd
import math
```

```python
#read census file
census = pd.read_csv("sample_census.csv")
census
```

```python
#clean the data - some rows have income as -666666666.0
census = census.loc[0:8]
census
```

```python
ssqd = (census["Median Age"] - 30)**2
census["Age_sqd"] = ssqd
#ssqd
census
```

```python
variance_age = census["Age_sqd"].sum()/len(census)
variance_age
```

```python
sd_age = math.sqrt(variance_age)
sd_age
```

```python
census["z_score_age"] = 1-abs((census["Median Age"] - 30)/sd_age)
#census.dtypes
census
```
